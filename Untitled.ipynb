```python
df=pd.read_csv('Breast cancer.txt',sep ='\t')
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Class'],axis=1)
y=df.Class
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_train,y_train)
```

```python
df['Bare Nuclei']=df['Bare Nuclei'].replace('?',0)
```

```python
clf.score(X_test,y_test)
```

```python
graph=Source(tree.export_graphviz(clf,out_file=None,feature_names=list(X),
                                 class_names=['Negative','Positive'],
                                 filled=True))
display(SVG(graph.pipe(format='svg')))
```

```python
max_depth=range(1,100)
```

```python
scores_data=pd.DataFrame()
```

```python
for i in max_depth:
    clf=tree.DecisionTreeClassifier(criterion='entropy',max_depth=i)
    clf.fit(X_train,y_train)
    train_score=clf.score(X_train,y_train)
    test_score=clf.score(X_test,y_test)
    mean_cross_val_score=cross_val_score(clf,X_train,y_train,cv=5).mean()
    temp_score=pd.DataFrame({'max_depth':[i], 
                             'train_score':[train_score],
                             'test_score':[test_score],
                            'cross_val_score':[mean_cross_val_score]})
    scores_data=scores_data.append(temp_score)
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
y_pred=best_clf.predict(X_test)
```

```python
precision_score(y_test,y_pred,average="binary", pos_label='malignant')
```

```python
recall_score(y_test,y_pred,average="binary", pos_label='malignant')
```

```python
y_pred_prob=best_clf.predict_proba(X_test)
```

```python
pd.Series(y_pred_prob[:,1]).hist()
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1],pos_label='malignant')
roc_auc= auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange',
          label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
